```python
%pylab inline
import os
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
amostra = np.loadtxt('./reward_file.dat')
amostra2 = np.loadtxt('./reward_file2.dat')
amostra3 = np.loadtxt('./reward_file3.dat') #5 action com bola laranja
amostra4 = np.loadtxt('./reward_file4.dat') #5 action com bola branca sem retorno
amostra5 = np.loadtxt('./reward_file5.dat') #5 action com bola branca com retorno
amostra6 = np.loadtxt('./reward_file6.dat') #5 action com bola branca com retorno
# amostra3 = np.loadtxt('./reward_7action.dat')
```

```python
print amostra[:,2]
```

```python
d = {'episodes' : pd.Series(amostra[:,1], index=amostra[:,0]), 
     'episode_rewards' : pd.Series(amostra[:,2], index=amostra[:,0])}
d2 = {'episodes' : pd.Series(amostra2[:,1], index=amostra2[:,0]), 
     'episode_rewards' : pd.Series(amostra2[:,2], index=amostra2[:,0])}
d3 = {'episodes' : pd.Series(amostra3[:,1], index=amostra3[:,0]), 
     'episode_rewards' : pd.Series(amostra3[:,2], index=amostra3[:,0])}
d4 = {'episodes' : pd.Series(amostra4[:,1], index=amostra4[:,0]), 
     'episode_rewards' : pd.Series(amostra4[:,2], index=amostra4[:,0])}
d5 = {'episodes' : pd.Series(amostra5[:,1], index=amostra5[:,0]), 
     'episode_rewards' : pd.Series(amostra5[:,2], index=amostra5[:,0])}
d6 = {'episodes' : pd.Series(amostra6[:,1], index=amostra6[:,0]), 
     'episode_rewards' : pd.Series(amostra6[:,2], index=amostra6[:,0])}
```

```python
df = pd.DataFrame(d)
df2 = pd.DataFrame(d2)
df3 = pd.DataFrame(d3)
df4 = pd.DataFrame(d4)
df5 = pd.DataFrame(d5)
df6 = pd.DataFrame(d6)
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.episode_rewards)
```

```python
plt.figure(figsize=(12,8))
plt.plot(df2.episode_rewards, label=u'5 acoes com bola laranja sem variar a cor e iluminacao do campo')
plt.plot(df3.episode_rewards, label=u'5 acoes com bola laranja sem retorno ao gol')
plt.plot(df4.episode_rewards, label=u'5 acoes com bola branca sem retorno ao gol')
plt.plot(df5.episode_rewards, 'c', label=u'5 acoes com bola branca com retorno ao gol')
plt.plot(df6.episode_rewards, 'y', label=u'5 acoes com bola branca com retorno ao gol')
plt.plot(df.episode_rewards, 'r', label=u'7 acoes com bola branca com retorno ao gol')
plt.legend()
plt.xlabel(u'Steps', fontsize=16)
plt.ylabel(u'Accumulated reward', fontsize=16)
# plt.yscale('log')
```

```python
plt.figure(figsize=(10,6))
plt.plot(df6.episode_rewards[df6.index>150], 'y')
plt.plot(df3.episode_rewards[df3.index>150])
plt.plot(df.episode_rewards[df.index>150], 'r')
# plt.yscale('log')
```
